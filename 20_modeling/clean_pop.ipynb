```python
population.sample(15)
```

```python
#Rename columns

population.rename(columns = {'CENSUS2010POP': 'POPESTIMATE2010'}, inplace=True)

```

```python
for col in population.columns:
    if col[:3] == "POP":
        population.rename(columns={col: col[-4:]}, inplace= True)

population.columns

```

```python
# melt the population fields
melted_pop = pd.melt(population, ["STATE", "COUNTY", "STNAME", "CTYNAME"])



# rename FIPS and Year Columns 

melted_pop.rename(columns = {"variable" : "Year", "value": "Population"}, inplace = True)

melted_pop.sample(14)
```

```python
## Adding FIPS Code to Population
melted_pop['STATE'] = melted_pop['STATE'].astype(str)
melted_pop['STATE1'] = melted_pop['STATE'].apply(lambda x: x.zfill(2))

melted_pop['COUNTY'] = melted_pop['COUNTY'].astype(str)
melted_pop['COUNTY1'] = melted_pop['COUNTY'].apply(lambda x: x.zfill(3))

melted_pop['fips'] = melted_pop['STATE1'] + melted_pop['COUNTY1']

melted_pop.head()
```

```python
## Save to parquet the cleaned file

melted_pop.to_parquet("/Users/preetkhowaja/Documents/MIDS Fall 2021/Practicing DS/pds2021-opioids-pds6/10_modified_data/pop_final.parquet", 
engine = 'fastparquet')
```
