```python
# Merge opioid and vital first

merge1 = pd.merge(opioid, vital, left_on=["fips", "year"], right_on=["County Code", "Year"], how='outer')

merge1.rename({"fips": "fips_opi", "County": "county_vital", "County Code": "fips_vital", "Year": "year_vital"}, axis=1, inplace=True)

merge1

# I tried to convert non-missing year values to integer but failed...
#merge1["year"] = np.where(merge1["year"].notnull(), merge1["year"].values.astype(int), merge1["year"])

```

```python
# Convert the fips column in merge1 data to strings and fill them to 5 digits. (for better merging with population data)

merge1["fips_opi"] = merge1["fips_opi"].astype(str)

merge1["fips_opi"] = merge1["fips_opi"].apply(lambda x: x.zfill(5))

#merge1 = merge1.drop(columns=["year", "County Code"])

# Convert the year column in population data to int64 so that the type matches

population["Year"] = population["Year"].astype(int)
```

```python
# Merge opioid and vital deaths data with population data, then

merge2 = pd.merge(merge1, population, left_on=["fips_opi", "year"], right_on=["FIPS", "Year"], how='outer')

merge2

#merge_final = merge2[["BUYER_STATE", "BUYER_COUNTY", "Year", "MME", "Deaths", "Population", "fips"]]

```
