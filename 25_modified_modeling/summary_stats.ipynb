## Stats for Florida

```python
opi["policy_year"] = np.where(opi["year"]<2010, 0, 1)

deaths["policy_year"] = np.where(deaths["Year"]<2010, 0, 1)
```

```python
control_fl = ["GA", "SC", "AL", "MS"]
control_fl_d = ["Georgia", "South Carolina", "Alabama", "Mississippi"]
```

```python
# Stats for control states before policy year

np.round(opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"]==0)][["MME", "MME per cap"]].describe(), 2)
```

```python
deaths[(deaths["STNAME"].isin(control_fl_d)) & (deaths["policy_year"]==0)][["Deaths", "deaths_per_cap"]].describe()
```

```python
# Stats for control states after policy year

np.round(opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"]==1)][["MME", "MME per cap"]].describe(), 2)
```

```python
deaths[(deaths["STNAME"].isin(control_fl_d)) & (deaths["policy_year"]==1)][["Deaths", "deaths_per_cap"]].describe()
```

```python
# Stats for Florida before policy year
np.round(opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"]==0)][["MME", "MME per cap"]].describe(), 2)
```

```python
deaths[(deaths["STNAME"]=="Florida") & (deaths["policy_year"]==0)][["Deaths", "deaths_per_cap"]].describe()
```

```python
# Stats for Florida after policy year
np.round(opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"]==1)][["MME", "MME per cap"]].describe(), 2)
```

```python
deaths[(deaths["STNAME"]=="Florida") & (deaths["policy_year"]==1)][["Deaths", "deaths_per_cap"]].describe()
```

```python
# difference in means

# before policy
diff1 = (
    opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 0)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 0)][
        "MME per cap"
    ].mean()
)

# after policy
diff2 = (
    opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 1)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 1)][
        "MME per cap"
    ].mean()
)

print(diff1, diff2)
```

```python
# difference in means

# change of control before and after policy
diff1 = (
    opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 1)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"].isin(control_fl)) & (opi["policy_year"] == 0)][
        "MME per cap"
    ].mean()
)

# change of Florida before and after policy
diff2 = (
    opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 1)]["MME per cap"].mean()
    - opi[(opi["BUYER_STATE"] == "FL") & (opi["policy_year"] == 0)]["MME per cap"].mean()
)

print(diff1, diff2)
```
