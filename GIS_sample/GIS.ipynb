## Exercise 4

Using sjoin_nearest, find the closest Whole Foods to every census block group. 

```python
# import the package that spatial join will need
import pygeos

# spatial join wholefoods to the census block data and create a new column of nearest distance
census_wholefoods = ny_census.sjoin_nearest(wholefoods, how='left', distance_col="distance_wholefoods")

census_wholefoods.head(5)
```

## Exercise 5

calculate the population density of each census block group and drop any census block groups with population densities below 100 people per square kilometer.

```python
# see what columns are in the dataset
census_wholefoods.columns
```

```python
# create a column of area
census_wholefoods["area"] = census_wholefoods.area

# convert square meter to square kilometer
census_wholefoods["area_km2"] = census_wholefoods["area"] / 1000000

# genearte population density
census_wholefoods["pop_density"] = census_wholefoods["population"] / census_wholefoods["area_km2"]

# keep census block groups with population densities no less than 100 people per square kilometer
census_wholefoods_urban = census_wholefoods[census_wholefoods["pop_density"] >= 100]
```

## Exercise 6

drop any block groups that are already less than 8 km from a Whole Foods.

```python
# keep blocks that are no less than 8 km from Whole Foods
census_wholefoods_near = census_wholefoods_urban[census_wholefoods_urban["distance_wholefoods"] >= 8000]

census_wholefoods_near = census_wholefoods_near.drop(["index_right"], axis='columns')
```

## Exercise 7

for each remaining block group, calculate the distance to the nearest NON-Whole Foods. 

```python
join = census_wholefoods_near.sjoin_nearest(competitors, how='left', distance_col='distance_competitors')

# rename the column after spatial join to avoid confusion
join.rename({"entity_nam_right": "competitor_name_1"}, axis='columns', inplace=True)
```

```python
# check if there's any duplicated value in terms of "STATEFP", "COUNTYFP", "TRACTCE", "BLKGRPCE"
join[join.duplicated(["STATEFP", "COUNTYFP", "TRACTCE", "BLKGRPCE"], keep=False)]
```

```python
# keep the first entry of the duplicates values
join_drop_dup = join.drop_duplicates(subset=["STATEFP", "COUNTYFP", "TRACTCE", "BLKGRPCE"], keep='first')

# check if there's any duplicated values 
assert len(join_drop_dup[join_drop_dup.duplicated(["STATEFP", "COUNTYFP", "TRACTCE", "BLKGRPCE"], keep=False)]) == 0
```

What does the distribution of those distances look like?

```python
# plot the distribution of the distance to competitors
join_drop_dup["distance_competitors"].hist()
```

The histogram plot shows that the majority of competitors are located within 10km of Whole Foods Market.

## Exercise 8

construct a variable the gives us the share of people over 25 in each Block Group that have at least a Bachelors Degree. 

```python
# create a column of the share of people over 25 in each Block Group that have at least a Bachelors Degree
join_drop_dup["share_bachelor_25"] = (
    join_drop_dup["ALWGE022"]
    + join_drop_dup["ALWGE023"]
    + join_drop_dup["ALWGE024"]
    + join_drop_dup["ALWGE025"]
) / join_drop_dup["ALWGE001"]
```

## Exercise 9

look for places where the share of people with college degrees is over 50%, median household income is over $90,000, and the nearest competitor’s grocery store is more than 8km away, and the nearest Whole Foods is over 8km away.

```python
# describe the median household income to get a basic sense
join_drop_dup["md_hh_inc"].describe()
```

```python
# pick the candidates that meet all the requirements
candidates = join_drop_dup[
    (join_drop_dup["share_bachelor_25"] > 0.5)
    & (join_drop_dup["md_hh_inc"] > 90000)
    & (join_drop_dup["distance_competitors"] > 8000)
    & (join_drop_dup["distance_wholefoods"] > 8000)
]
```

Using these filters, can you identify any counties that seem like especially good candidates (e.g. counties with a fair number of people in census block groups that fit our criteria?)

```python
# We think counties exceed the median population should be considered as good candidates
# so we pick counties that has above-median population

pop_median = candidates["population"].median()

candidates[candidates["population"] >= pop_median]["COUNTY"].value_counts()
```

Based on the criterion above, there are 13 good candidates in total. They're located in Westchester County, Erie County, Niagara County, Onondaga County, Oswego County, Putnam County, and Suffolk County.

## Exercise 10

buffer your block groups by 15 kilometers and set the new buffered polygons as your GeoDataFrame’s “offical geometry”

```python
# create buffer of 15km using the cesus_wholefoods dataframe
join["buffers"] = join.buffer(15000)

# reset the geometry as buffers
wholefoods_buffers = join.set_geometry("buffers")
```

```python
# plotting
wholefoods_buffers.plot()
```

```python
# drop index right
wholefoods_buffers = wholefoods_buffers.drop(["index_right"], axis='columns')
```

## Exercise 11

merge your buffered polygons with your GeoDataFrame of other grocers.

```python
# merge the buffered polygons with other grocers
joined = wholefoods_buffers.sjoin(competitors[["entity_nam", "geometry"]])

# rename the name of other grocers as "compettitor_name_2" to avoid confusion
joined.rename({"entity_nam": "competitor_name_2"}, axis='columns', inplace=True)
```

## Exercise 12

groupby our data to get the COUNT of competitors within 15km.

```python
# groupby on block-group level and count the number of competitors
joined["competitors_count"] = joined.groupby(["STATEFP", "COUNTYFP", "TRACTCE", "BLKGRPCE"])["competitor_name_2"].transform("count")

joined.head(15)
```

## Exercise 13

let’s try using the same filter we used before, but this time subsetting for census block groups with less than three competitors within 15km and no competitors within 5km.

```python
# create a column of the share of people over 25 in each Block Group that have at least a Bachelors Degree
joined["share_bachelor_25"] = (
    joined["ALWGE022"]
    + joined["ALWGE023"]
    + joined["ALWGE024"]
    + joined["ALWGE025"]
) / joined["ALWGE001"]
```
